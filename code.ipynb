```python
import pandas as pd
import numpy as np
```

```python
housing = pd.read_csv("Zip_zhvi_uc_sfrcondo_tier_0.33_0.67_sm_sa_mon.csv")
housing = housing[housing['State'] == 'CA']
housing_county = housing[['RegionName', 'RegionType', 
                          'State', 'City', 'CountyName', 
                          '2019-01-31']]
#housing_county_mean = housing_county_mean.rename(columns={"2019-01-31": "Mean House Price 2019-01-31"})[['Mean House Price 2019-01-31']]
#housing_county_mean= housing_county_mean.reset_index()
#housing_county_mean.head()
housing_county
```

```python
housing_scores = pd.read_csv('public_housing_physical_inspection_scores_grouped.csv')
housing_scores['COUNTY'] = housing_scores['COUNTY'] + ' County'
housing_scores
```

```python
housing_county1 = housing_county.merge(housing_scores, left_on = "CountyName", right_on = 'COUNTY').drop("COUNTY", 1)
housing_county1
```

```python
transport = pd.read_csv("County_Transportation_Profiles.csv")
transport = transport[transport['State Name'] == 'California']
transport = transport[['County Name', 'State Name', 
                       'Percent of resident workers who commute by transit', 
                       'Number of business establishments']]
transport.head()
```

```python
#rates per 100,000
crime_rates = pd.read_csv("crime_rates.csv")
crime_rates = crime_rates.iloc[2:]
crime_rates.columns = crime_rates.iloc[0]
crime_rates = crime_rates.iloc[1:]
crime_rates = crime_rates[['County', 'Violent', 'Murder', 'Robbery', 'Aggrevated assault']]
counties_cr = crime_rates['County'] + " County"
crime_rates['County'] = counties_cr
crime_rates = crime_rates.dropna()
crime_rates.head()
```

```python
pollution = pd.read_csv("pollution.csv")
pollution = pollution[['California County', 'ZIP', 'CES 3.0 Score',' CES 3.0 Percentile', 'Poverty', 'Unemployment']]
counties_pol = pollution['California County'] + ' County'
pollution['California County'] = counties_pol
#pollution = pollution.groupby('California County').mean()[["CES 3.0 Score",
                                                           #"Poverty","Unemployment"]]
#pollution = pollution.reset_index()
pollution.head()
```

```python
scores_entities = pd.read_csv("scores_ca2019_entities.csv", encoding='latin-1')
scores_entities.head()
```

```python
scores = pd.read_csv("scores_ca2019.csv")
scores = scores[['County Code', 'Grade', 'Mean Scale Score']].dropna()
scores['Mean Scale Score'] = pd.to_numeric(scores['Mean Scale Score'],errors='coerce')
scores_grouped = scores.groupby(['County Code', 'Grade']).mean()
scores_grouped = scores_grouped.reset_index()
```

```python
new_scores = scores_grouped.merge(scores_entities, how = 'right', left_on='County Code', right_on = 'County Code')
new_scores = new_scores[['County Code', 'Mean Scale Score', 'County Name']][7:]
counties_scores = new_scores['County Name'] + ' County'
new_scores['County Name'] = counties_scores
new_scores
```

```python
housing_pollution = housing_county1.merge(pollution, how = 'right', left_on = 'RegionName', right_on = 'ZIP')
housing_pollution = housing_pollution.dropna().drop(['California County',
                                                     'ZIP'], 1).rename(columns={'2019-01-31': 
                                                                                'House Price 01-2019'})
housing_pollution.head()
```

```python
ho_po_transport = housing_pollution.merge(transport, how='right', left_on = 'CountyName', right_on = 'County Name')
ho_po_transport = ho_po_transport.dropna().drop(['CountyName', 'State Name'], 1)
ho_po_transport.head()
```

```python
ho_po_tr_crimes = ho_po_transport.merge(crime_rates, how = 'right', left_on = 'County Name', right_on = "County")
ho_po_tr_crimes.head()
```

```python
ho_po_tr_cr_scores = ho_po_tr_crimes.merge(new_scores, how = 'right', left_on = 'County Name', right_on = "County Name")
```

```python
ho_po_tr_cr_scores = ho_po_tr_cr_scores.drop_duplicates()
ho_po_tr_cr_scores.to_csv('all.csv')
```

```python
alli = ho_po_tr_cr_scores.drop(['RegionType', 'State', 'City', ' CES 3.0 Percentile', 
                         'Poverty', 'Unemployment', 'Number of business establishments', 
                         'County', 'Murder', 'Robbery', 'Aggrevated assault', 'County Code'], axis=1)
alli = alli.dropna()
alli['Violent'] = alli['Violent'].astype('float')
alli
```

```python
all_df = alli.groupby('RegionName', as_index = False).agg({'House Price 01-2019': 'mean', 
                                                           'AVERAGE_INSPECTION_SCORE': 'mean', 
                                                           'CES 3.0 Score': 'mean', 'County Name': 'first', 
                                                           'Percent of resident workers who commute by transit': 'mean',
                                                           'Violent': 'mean', 'Mean Scale Score': 'mean'})
all_df
```
